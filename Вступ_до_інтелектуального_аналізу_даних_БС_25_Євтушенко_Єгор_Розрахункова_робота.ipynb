##–ö—Ä–∏—Ç–µ—Ä—ñ—ó –æ—Ü—ñ–Ω–∫–∏ —è–∫–æ—Å—Ç—ñ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü—ñ—ó (—Ñ–æ—Ä–º—É–ª–∏ 2.23-2.28)

####1. –í–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è ($W_k$)

–í–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è –≤–∏–º—ñ—Ä—é—î —â—ñ–ª—å–Ω—ñ—Å—Ç—å –æ–±'—î–∫—Ç—ñ–≤ —É –∫–ª–∞—Å—Ç–µ—Ä—ñ:


$W_k = \sum_{q=1}^K \sum_{x \in C_q} \|x - \mu_q\|^2,$

–¥–µ:
- $x$ ‚Äî –æ–±'—î–∫—Ç–∏ –≤ –∫–ª–∞—Å—Ç–µ—Ä—ñ $C_q$;
- $\mu_q$ ‚Äî —Ü–µ–Ω—Ç—Ä –∫–ª–∞—Å—Ç–µ—Ä–∞ $q$.



####2. –ú—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è ($B_k$)

–ú—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è –ø–æ–∫–∞–∑—É—î –≤—ñ–¥—Å—Ç–∞–Ω—å –º—ñ–∂ –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏:

$B_k = \sum_{q=1}^K n_q \cdot \|\mu_q - \mu\|^2,$

–¥–µ:
- $n_q$ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–±'—î–∫—Ç—ñ–≤ —É –∫–ª–∞—Å—Ç–µ—Ä—ñ $q$;
- $\mu$ ‚Äî —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –≤—Å—ñ—Ö –¥–∞–Ω–∏—Ö.



####3. –ö—Ä–∏—Ç–µ—Ä—ñ–π —Å–∏–ª—É–µ—Ç—É ($S_k$)

–°–∏–ª—É–µ—Ç –≤–∏–∑–Ω–∞—á–∞—î, –Ω–∞—Å–∫—ñ–ª—å–∫–∏ –¥–æ–±—Ä–µ –æ–±'—î–∫—Ç –Ω–∞–ª–µ–∂–∏—Ç—å —Å–≤–æ—î–º—É –∫–ª–∞—Å—Ç–µ—Ä—É –≤ –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—ñ –∑ —ñ–Ω—à–∏–º–∏:

$S_k = \frac{b - a}{\max(a, b)}$,


  –¥–µ:
  - $a$ ‚Äî —Å–µ—Ä–µ–¥–Ω—è –≤—ñ–¥—Å—Ç–∞–Ω—å –¥–æ –æ–±'—î–∫—Ç—ñ–≤ —É —Å–≤–æ—î–º—É –∫–ª–∞—Å—Ç–µ—Ä—ñ;
  - $b$ ‚Äî —Å–µ—Ä–µ–¥–Ω—è –≤—ñ–¥—Å—Ç–∞–Ω—å –¥–æ –æ–±'—î–∫—Ç—ñ–≤ —É –Ω–∞–π–±–ª–∏–∂—á–æ–º—É —ñ–Ω—à–æ–º—É –∫–ª–∞—Å—Ç–µ—Ä—ñ.



####4. –ó–∞–≥–∞–ª—å–Ω–∞ –æ—Ü—ñ–Ω–∫–∞ ($J_k$)

–û—Ü—ñ–Ω–∫–∞ —è–∫–æ—Å—Ç—ñ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü—ñ—ó:
$J_k = W_k + B_k.$


###–î–æ–¥–∞—Ç–∫–æ–≤—ñ —Ñ–æ—Ä–º—É–ª–∏
$
\min_{K, x_i \in X_j} \frac{1}{K} \sum_{q=1}^K \frac{1}{n_q} \sum_{i=1}^{n_q} \sum_{p=1}^{18} \left( x_{ip} - \bar{x}_p^q \right)^2
\Bigg/
\left(

\right)
$
$
\min_{K, x_i \in X_j} \frac{1}{K} \sum_{q=1}^K \frac{1}{n_q} \sum_{i=1}^{n_q}
\sqrt{\sum_{p=1}^{18} \left( x_{ip} - \bar{x}_p^q \right)^2}
\Bigg/
\left(
\frac{1}{C_K^2} \sum_{i=1}^K \sum_{j=i+1}^K
\sqrt{\sum_{p=1}^{18} \left( \bar{x}_p^i - \bar{x}_p^j \right)^2}
\right)
$

–ü–æ—è—Å–Ω–µ–Ω–Ω—è:

- $K$ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∫–ª–∞—Å—Ç–µ—Ä—ñ–≤.
- $n_q$ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–±'—î–∫—Ç—ñ–≤ —É –∫–ª–∞—Å—Ç–µ—Ä—ñ $q$.
- $x_{ip}$ ‚Äî –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ –æ–±'—î–∫—Ç–∞ $i$ —É $p$-–º—É –≤–∏–º—ñ—Ä—ñ.
- $\overline{x}_{p}^q$ ‚Äî —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∏ $p$ —É –∫–ª–∞—Å—Ç–µ—Ä—ñ $q$.
- $C_K^2$ ‚Äî –∫–æ–º–±—ñ–Ω–∞—Ü—ñ–π–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–∞—Ä –∫–ª–∞—Å—Ç–µ—Ä—ñ–≤, –æ–±—á–∏—Å–ª—é—î—Ç—å—Å—è –∑–∞ —Ñ–æ—Ä–º—É–ª–æ—é:
$C_K^2 = \frac{K(K-1)}{2}.$

## –ó–∞–≤–¥–∞–Ω–Ω—è A

```python
# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É –≤–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ—ó –¥–∏—Å–ø–µ—Ä—Å—ñ—ó
def calculate_within_cluster_dispersion(data, cluster_column, value_columns):
    dispersion = 0
    for cluster in data[cluster_column].unique():
        cluster_data = data[data[cluster_column] == cluster][value_columns]
        cluster_center = cluster_data.mean(axis=0)
        dispersion += ((cluster_data - cluster_center) ** 2).sum(axis=0).sum()
    return dispersion

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É –º—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ—ó –¥–∏—Å–ø–µ—Ä—Å—ñ—ó
def calculate_between_cluster_dispersion(data, cluster_column, value_columns):
    overall_mean = data[value_columns].mean(axis=0)
    dispersion = 0
    for cluster in data[cluster_column].unique():
        cluster_data = data[data[cluster_column] == cluster][value_columns]
        cluster_center = cluster_data.mean(axis=0)
        cluster_size = cluster_data.shape[0]
        dispersion += cluster_size * ((cluster_center - overall_mean) ** 2).sum()
    return dispersion

# –§—É–Ω–∫—Ü—ñ—è –¥–ª—è –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –∫—Ä–∏—Ç–µ—Ä—ñ—é —Å–∏–ª—É–µ—Ç—É
def calculate_silhouette_score(data, cluster_labels):
    return silhouette_score(data, cluster_labels)

# –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –∫—Ä–∏—Ç–µ—Ä—ñ—ó–≤ –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –≤–∞—Ä—ñ–∞–Ω—Ç—É –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü—ñ—ó
criteria_results = []

for data, cluster_column in zip([clusters_8, clusters_12, clusters_14], ["Cl8_APHR", "Cl12_APHR", "Cl14_APHR"]):
    # –í–∏–±—ñ—Ä —á–∏—Å–ª–æ–≤–∏—Ö –∫–æ–ª–æ–Ω–æ–∫ –¥–ª—è –∞–Ω–∞–ª—ñ–∑—É
    value_columns = data.select_dtypes(include=["number"]).columns.tolist()

    # –û–±—á–∏—Å–ª–µ–Ω–Ω—è –∫—Ä–∏—Ç–µ—Ä—ñ—ó–≤
    W_k = calculate_within_cluster_dispersion(data, cluster_column, value_columns)
    B_k = calculate_between_cluster_dispersion(data, cluster_column, value_columns)
    S_k = calculate_silhouette_score(data[value_columns], data[cluster_column])
    J_k = W_k + B_k

    # –î–æ–¥–∞–≤–∞–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤ –¥–æ —Ç–∞–±–ª–∏—Ü—ñ
    criteria_results.append({
        "–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∫–ª–∞—Å—Ç–µ—Ä—ñ–≤": cluster_column,
        "–í–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è": round(W_k, 2),
        "–ú—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è": round(B_k, 2),
        "–ö—Ä–∏—Ç–µ—Ä—ñ–π —Å–∏–ª—É–µ—Ç—É": round(S_k, 2),
        "–ó–∞–≥–∞–ª—å–Ω–∞ –æ—Ü—ñ–Ω–∫–∞": round(J_k, 2),
    })

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–∞–±–ª–∏—Ü—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
criteria_df = pd.DataFrame(criteria_results)


# –§–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è —á–∏—Å–µ–ª –¥–ª—è –∑—Ä—É—á–Ω–æ–≥–æ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è
pd.options.display.float_format = '{:,.2f}'.format

try:
    from IPython.display import display
    display(criteria_df.style.set_properties(**{
        'background-color': '#f9f9f9',
        'color': 'black',
        'border-color': 'black',
        'text-align': 'center'
    }).highlight_max(subset=["–ú—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è", "–ö—Ä–∏—Ç–µ—Ä—ñ–π —Å–∏–ª—É–µ—Ç—É"], color='lightgreen')
      .highlight_min(subset=["–í–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è"], color='lightpink'))
except ImportError:
    print(criteria_df)
```

### –°–ø—Ä–æ–±–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ —ñ–Ω—à–∏–π –º–µ—Ç–æ–¥ —ñ–Ω—ñ—Ü—ñ–∞–ª—ñ–∑–∞—Ü—ñ—ó —Ü–µ–Ω—Ç—Ä—ñ–≤

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
url_8_clusters = "https://raw.githubusercontent.com/yeyevtushenko/IntroductionIAD/refs/heads/Data/8%20%D0%BA%D0%BB%D0%B0%D1%81%D1%82%D0%B5%D1%80%D1%96%D0%B2.csv"
url_12_clusters = "https://raw.githubusercontent.com/yeyevtushenko/IntroductionIAD/refs/heads/Data/12%20%D0%BA%D0%BB%D0%B0%D1%81%D1%82%D0%B5%D1%80%D1%96%D0%B2.csv"
url_14_clusters = "https://raw.githubusercontent.com/yeyevtushenko/IntroductionIAD/refs/heads/Data/14%20%D0%BA%D0%BB%D0%B0%D1%81%D1%82%D0%B5%D1%80%D1%96%D0%B2.csv"

try:
    clusters_8 = pd.read_csv(url_8_clusters)
    clusters_12 = pd.read_csv(url_12_clusters)
    clusters_14 = pd.read_csv(url_14_clusters)

    print("–î–∞–Ω—ñ —É—Å–ø—ñ—à–Ω–æ –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–æ!")
except Exception as e:
    print(f"–ü–æ–º–∏–ª–∫–∞ –ø—Ä–∏ –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—ñ –¥–∞–Ω–∏—Ö: {e}")
```

```python
# –í–∏–±—ñ—Ä —á–∏—Å–ª–æ–≤–∏—Ö –∫–æ–ª–æ–Ω–æ–∫ –¥–ª—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü—ñ—ó
value_columns = [col for col in clusters_8.columns if clusters_8[col].dtype in [np.float64, np.int64]]

# –í–∏–∫–æ–Ω—É—î–º–æ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü—ñ—é –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –º–µ—Ç–æ–¥—É k-means++
n_clusters_list = [8, 12, 14]
results = []

for n_clusters, data in zip(n_clusters_list, [clusters_8, clusters_12, clusters_14]):
    # –í–∏–∫–æ–Ω—É—î–º–æ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü—ñ—é
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
    cluster_labels = kmeans.fit_predict(data[value_columns])

    # –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –≤–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ—ó –¥–∏—Å–ø–µ—Ä—Å—ñ—ó (W_k)
    W_k = kmeans.inertia_

    # –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –º—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ—ó –¥–∏—Å–ø–µ—Ä—Å—ñ—ó (B_k)
    overall_mean = data[value_columns].mean().values
    B_k = 0
    for i in range(n_clusters):
        cluster_data = data[value_columns][cluster_labels == i]
        cluster_center = kmeans.cluster_centers_[i]
        cluster_size = len(cluster_data)
        B_k += cluster_size * np.sum((cluster_center - overall_mean) ** 2)

    # –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –∫—Ä–∏—Ç–µ—Ä—ñ—é —Å–∏–ª—É–µ—Ç—É (S_k)
    silhouette_avg = silhouette_score(data[value_columns], cluster_labels)

    # –ó–±–µ—Ä—ñ–≥–∞—î–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
    results.append({
        "–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∫–ª–∞—Å—Ç–µ—Ä—ñ–≤": n_clusters,
        "–í–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è": round(W_k, 2),
        "–ú—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è": round(B_k, 2),
        "–ö—Ä–∏—Ç–µ—Ä—ñ–π —Å–∏–ª—É–µ—Ç—É": round(silhouette_avg, 2)
    })

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–∞–±–ª–∏—Ü—ñ –∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
criteria_df = pd.DataFrame(results)

# –§–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è —á–∏—Å–µ–ª –¥–ª—è –∑—Ä—É—á–Ω–æ–≥–æ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è
pd.options.display.float_format = '{:,.2f}'.format

# –í–∏–≤–µ–¥–µ–Ω–Ω—è—Ç–∞–±–ª–∏—Ü—ñ
print("–†–µ–∑—É–ª—å—Ç–∞—Ç–∏ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –∫—Ä–∏—Ç–µ—Ä—ñ—ó–≤ –¥–ª—è –∫–ª–∞—Å—Ç–µ—Ä—ñ–≤:")

try:
    from IPython.display import display
    display(criteria_df.style.set_properties(**{
        'background-color': '#f9f9f9',
        'color': 'black',
        'border-color': 'black',
        'text-align': 'center'
    }).highlight_max(subset=["–ú—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è", "–ö—Ä–∏—Ç–µ—Ä—ñ–π —Å–∏–ª—É–µ—Ç—É"], color='lightgreen')
      .highlight_min(subset=["–í–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è"], color='lightpink'))
except ImportError:
    print(criteria_df)
```

```python
url_8_clusters = "https://raw.githubusercontent.com/yeyevtushenko/IntroductionIAD/refs/heads/Data/8%20%D0%BA%D0%BB%D0%B0%D1%81%D1%82%D0%B5%D1%80%D1%96%D0%B2.csv"
url_12_clusters = "https://raw.githubusercontent.com/yeyevtushenko/IntroductionIAD/refs/heads/Data/12%20%D0%BA%D0%BB%D0%B0%D1%81%D1%82%D0%B5%D1%80%D1%96%D0%B2.csv"
url_14_clusters = "https://raw.githubusercontent.com/yeyevtushenko/IntroductionIAD/refs/heads/Data/14%20%D0%BA%D0%BB%D0%B0%D1%81%D1%82%D0%B5%D1%80%D1%96%D0%B2.csv"

try:
    clusters_8 = pd.read_csv(url_8_clusters)
    clusters_12 = pd.read_csv(url_12_clusters)
    clusters_14 = pd.read_csv(url_14_clusters)

    print("–î–∞–Ω—ñ —É—Å–ø—ñ—à–Ω–æ –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–æ!")
except Exception as e:
    print(f"–ü–æ–º–∏–ª–∫–∞ –ø—Ä–∏ –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—ñ –¥–∞–Ω–∏—Ö: {e}")

# –í–∏–¥—ñ–ª–µ–Ω–Ω—è –ø–æ—Ç—Ä—ñ–±–Ω–∏—Ö –∑–º—ñ–Ω–Ω–∏—Ö (—Ç—Ä—ñ–π–∫–∏: –ê–¢–°0, –ê–¢–î0, –ü—É–ª—å—Å0, –ê–¢–°1, –ê–¢–î1, –ü—É–ª—å—Å1, –ê–¢–°3, –ê–¢–î3, –ü—É–ª—å—Å3, –ê–¢–°4, –ê–¢–î4, –ü—É–ª—å—Å4)
selected_columns = ['–ê–¢–°0', '–ê–¢–î0', '–ü—É–ª—å—Å0', '–ê–¢–°1', '–ê–¢–î1', '–ü—É–ª—å—Å1', '–ê–¢–°3', '–ê–¢–î3', '–ü—É–ª—å—Å3', '–ê–¢–°4', '–ê–¢–î4', '–ü—É–ª—å—Å4']

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –ø—É—Å—Ç–æ–≥–æ —Å–ø–∏—Å–∫—É –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
results = []

# –ú–∞—Å—à—Ç–∞–±—É–≤–∞–Ω–Ω—è —Ç–∞ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü—ñ—è –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –∑ –Ω–∞–±–æ—Ä—ñ–≤ –¥–∞–Ω–∏—Ö
for clusters_data, n_clusters in zip([clusters_8, clusters_12, clusters_14], [8, 12, 14]):
    # –í–∏–¥—ñ–ª–µ–Ω–Ω—è –∑–º—ñ–Ω–Ω–∏—Ö
    selected_data = clusters_data[selected_columns]

    # –ú–∞—Å—à—Ç–∞–±—É–≤–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(selected_data)

    # –í–∏–∫–æ–Ω–∞–Ω–Ω—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü—ñ—ó –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –º–µ—Ç–æ–¥—É k-means++
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_data)

    # –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –≤–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ—ó –¥–∏—Å–ø–µ—Ä—Å—ñ—ó (W_k)
    W_k = kmeans.inertia_

    # –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –º—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–æ—ó –¥–∏—Å–ø–µ—Ä—Å—ñ—ó (B_k)
    overall_mean = scaled_data.mean(axis=0)
    B_k = 0
    for i in range(n_clusters):
        cluster_data = scaled_data[cluster_labels == i]
        cluster_center = kmeans.cluster_centers_[i]
        cluster_size = len(cluster_data)
        B_k += cluster_size * np.sum((cluster_center - overall_mean) ** 2)

    # –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ –∫—Ä–∏—Ç–µ—Ä—ñ—é —Å–∏–ª—É–µ—Ç—É (S_k)
    silhouette_avg = silhouette_score(scaled_data, cluster_labels)

    # –î–æ–¥–∞–≤–∞–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤ –¥–æ —Å–ø–∏—Å–∫—É
    results.append({
        "–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∫–ª–∞—Å—Ç–µ—Ä—ñ–≤": n_clusters,
        "–í–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è": round(W_k, 2),
        "–ú—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è": round(B_k, 2),
        "–ö—Ä–∏—Ç–µ—Ä—ñ–π —Å–∏–ª—É–µ—Ç—É": round(silhouette_avg, 2)
    })

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–∞–±–ª–∏—Ü—ñ –∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
criteria_df = pd.DataFrame(results)

# –§–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è —á–∏—Å–µ–ª –¥–ª—è –∑—Ä—É—á–Ω–æ–≥–æ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è
pd.options.display.float_format = '{:,.2f}'.format

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ç–∞–±–ª–∏—Ü—ñ –∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
print("–†–µ–∑—É–ª—å—Ç–∞—Ç–∏ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –∫—Ä–∏—Ç–µ—Ä—ñ—ó–≤ –¥–ª—è –∫–ª–∞—Å—Ç–µ—Ä—ñ–≤ (–¥–ª—è –æ–±—Ä–∞–Ω–∏—Ö —Ç—Ä—ñ–π–æ–∫ –∑–º—ñ–Ω–Ω–∏—Ö):")
try:
    from IPython.display import display
    display(criteria_df.style.set_properties(**{
        'background-color': '#f9f9f9',
        'color': 'black',
        'border-color': 'black',
        'text-align': 'center'
    }).highlight_max(subset=["–ú—ñ–∂–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è", "–ö—Ä–∏—Ç–µ—Ä—ñ–π —Å–∏–ª—É–µ—Ç—É"], color='lightgreen')
      .highlight_min(subset=["–í–Ω—É—Ç—Ä—ñ—à–Ω—å–æ–∫–ª–∞—Å—Ç–µ—Ä–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å—ñ—è"], color='lightpink'))
except ImportError:
    print(criteria_df)
```
